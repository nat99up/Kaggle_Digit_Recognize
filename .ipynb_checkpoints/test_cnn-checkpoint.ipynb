```python
## Loading the Training & Testing Data

import csv
import numpy as np


def trainLoading(file_name):
    
    file_contents = []
    labels = []
    datas = []
    
    
    with open(file_name) as file :
        
        lines = csv.reader(file)
        
        for line in lines:
            
            file_contents.append(line)
    
    file_contents.remove(file_contents[0])
    
    for line in file_contents:
        
        labels.append(labelArray(line[0]))
        datas.append(list(map(binarizeChar,line[1:])))
    
    labels = np.array(labels)
    datas = np.array(datas)
    
    return (labels,datas)
    
    
def testLoading(file_name):
    
    file_contents = []
    datas = []
    with open(file_name) as file :
        
        lines = csv.reader(file)
        
        for line in lines:
            
            file_contents.append(line)
    
    file_contents.remove(file_contents[0])
    
    for line in file_contents:

        datas.append(list(map(binarizeChar,line)))
        
    datas = np.array(datas)
    
    return datas


def binarizeChar(c):
    if c == '0':
        return 0
    else :
        return 1

def labelArray(l):
    arr = [0]*10
    arr[int(l)] = 1
    return arr
    
    
    
if __name__ == '__main__':
    
    (l,d) = trainLoading('train.csv')
    test_data = testLoading('test.csv')
    
```

```python
import tensorflow as tf
import random 

## Delete the current graph
tf.reset_default_graph()


def CNN(x):
    
    input_layer = tf.reshape(x,[-1,28,28,1])
    
    conv_1 = tf.layers.conv2d(inputs=input_layer,filters=16,kernel_size=[5,5],padding='same',name='Conv_Layer_1',activation=tf.nn.relu)
    
    conv_2 = tf.layers.conv2d(inputs=conv_1,filters=16,kernel_size=[5,5],padding='same',name='Conv_Layer_2',activation=tf.nn.relu)
    
    max_pool_1 = tf.layers.max_pooling2d(inputs=conv_2,pool_size=[2,2],strides=2,name='Max_Pool_1')
    
    conv_3 = tf.layers.conv2d(inputs=max_pool_1,filters=32,kernel_size=[3,3],padding='same',name='Conv_Layer_3',activation=tf.nn.relu)
    
    conv_4 = tf.layers.conv2d(inputs=conv_3,filters=32,kernel_size=[3,3],padding='same',name='Conv_Layer_4',activation=tf.nn.relu)
    
    max_pool_2 = tf.layers.max_pooling2d(inputs=conv_4,pool_size=[2,2],strides=2,name='Max_Pool_2')
    
    flat = tf.reshape(max_pool_2,[-1,7*7*32])
    
    dense_layer = tf.layers.dense(inputs=flat,units=1024,activation=tf.nn.relu)
    
    dropout = tf.layers.dropout(inputs=dense_layer, rate=0.4, training=tf_is_training)
    
    output_layer = tf.layers.dense(inputs=dropout,units=10)
    
    
    
    return output_layer
    

x = tf.placeholder(tf.float32,[None,784])
y_ = tf.placeholder(tf.float32,[None,10])
tf_is_training = tf.placeholder(tf.bool, [])

logits = CNN(x)

y = tf.nn.softmax(logits)

cross_entropy = tf.nn.softmax_cross_entropy_with_logits_v2(logits=logits,labels=y_)

#optimizer = tf.train.AdamOptimizer(learning_rate=0.003)

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.003)

train_step = optimizer.minimize(cross_entropy)

predict = tf.argmax(y,axis=1)

estimate = tf.equal(predict,tf.argmax(y_,axis=1))

accuarcy = tf.reduce_mean(tf.cast(estimate,tf.float32))

init = tf.global_variables_initializer()

with tf.Session() as sess:
    
    sess.run(init)
    
    epoch = 12600
    batch_size = 30
    s = 0
    
    
    for i in range(epoch):
        
        sess.run(train_step,feed_dict={x:d[s:s+batch_size],y_:l[s:s+batch_size],tf_is_training:True})
        
        s = (s + batch_size)%(42000-batch_size)
        
        if i % (epoch/40) == 0:
        
            rand_list = np.sort(random.sample(range(42000),1000))
            
            acc = sess.run(accuarcy,feed_dict={x:d[rand_list],y_:l[rand_list],tf_is_training:False})

            print(acc)
            
            
            
            if(acc>0.999):
                
                print('I got a 99.9% model !!')
                
                break
                
    
    print('start to process testing data')
    
    s = 0
    submit_prediction = np.array([])

    for j in range(28):

        submit_prediction_part = sess.run(tf.cast(predict,tf.int32),feed_dict={x:test_data[s:s+1000],tf_is_training:False})
        submit_prediction = np.concatenate((submit_prediction,submit_prediction_part))

        s = s + 1000

    submit_prediction = np.array(list(map(int,submit_prediction)))
    
    print('Done')
    
    
```

```python
#result寫入csv檔

with open('submission.csv','w',newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['ImageId','Label'])
    cnt = 1
    for i in submit_prediction:
        writer.writerow([cnt,i])
        cnt = cnt+1
```
